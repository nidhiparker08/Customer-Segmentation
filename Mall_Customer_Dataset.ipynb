```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Data raw\Mall_Customers.csv")
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

wcss = []
for k in range(1, 11):
    wcss.append(KMeans(n_clusters=k).fit(X).inertia_)

plt.plot(range(1, 11), wcss)
plt.xlabel("K")
plt.ylabel("WCSS")
plt.show()
```
